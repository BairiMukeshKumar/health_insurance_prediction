```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_excel("premiums_young_with_gr.xlsx")
df.head()
```

```python
df.columns=df.columns.str.replace(' ','_').str.lower()
df.head(1)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df[df.number_of_dependants<0].shape
```

```python
df['number_of_dependants']=abs(df['number_of_dependants'])
df['number_of_dependants'].describe()
```

```python
df[df.age>100].shape
```

```python
df1=df[df.age<=100].copy()
df1.describe()
```

```python
df1[df1.income_lakhs>100].shape
```

```python
df2=df1[df1.income_lakhs<=100].copy()
df2.describe()
```

```python
categorical_col=['gender','region', 'marital_status','bmi_category', 'smoking_status',
                 'employment_status', 'income_level', 'medical_history', 'insurance_plan']
for col in categorical_col:
    print(col,':',df2[col].unique())
```

```python
df2['smoking_status'].replace({'Smoking=0':'No Smoking',
                               'Does Not Smoke':'No Smoking',
                               'Not Smoking':'No Smoking'},inplace=True)
df2['smoking_status'].unique()
```

```python
risk_scores={
              'diabetes':6,
              'high blood pressure':6,
              'thyroid':5,
              'heart disease':8,
              'no disease':0,
              'none':0
            }
```

```python
df2[['disease1','disease2']]=df2['medical_history'].str.split('&',expand=True).apply(lambda x: x.str.lower().str.strip())
df2.fillna({'disease1':'none'},inplace=True)
df2.fillna({'disease2':'none'},inplace=True)
df2['total_risk_factors']=0
df2.head()
```

```python
for i in ['disease1','disease2']:
    df2['total_risk_factors']+=df2[i].map(risk_scores)
max_score=df2['total_risk_factors'].max()
min_score=df2['total_risk_factors'].min()
df2['noramlized_score']=(df2['total_risk_factors']-min_score)/(max_score-min_score)
df2.head()
```

```python
df2.insurance_plan.unique()
```

```python
df2['insurance_plan']=df2['insurance_plan'].map({'Bronze':1,'Silver':2,'Gold':3})
df2.insurance_plan.unique()
```

```python
df2.income_level.unique()
```

```python
df2['income_level']=df2['income_level'].map({'<10L':1,'10L - 25L':2,'25L - 40L':3,'> 40L':4})
df2.head()
```

```python
nominal_columns=['gender','region','marital_status','bmi_category','smoking_status','employment_status']
df3=pd.get_dummies(df2,columns=nominal_columns,drop_first=True,dtype=int)
df3.head()
```

```python
df4=df3.drop(['medical_history','disease1','disease2','total_risk_factors'],axis=1)
df4.head()
```

```python
df5=df4.drop(['income_level',
         'noramlized_score',
         'gender_Male',
         'region_Northwest',
         'region_Southeast',	
         'region_Southwest',
         'marital_status_Unmarried',
         'bmi_category_Obesity',
         'bmi_category_Overweight',	
         'bmi_category_Underweight',
         'smoking_status_Occasional',	
         'smoking_status_Regular',
         'employment_status_Salaried',	
         'employment_status_Self-Employed'],axis=1)
df5.head()
```
